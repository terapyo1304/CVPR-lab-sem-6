# 1. Reading,Displaying and Writing an image

```python
img = cv2.imread('king.jpeg')

# Displaying the image using Matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Normal Image')
plt.axis('off')
plt.show()

# Writing the image
if not os.path.exists('output'):
    os.makedirs('output')
cv2.imwrite(os.path.join('output', 'image.png'), img)
```

# 3. Resizing the Image

```python
img_resized = cv2.resize(img, (1050, 1000))
plt.imshow(cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.axis('off')
plt.show()
```

# 4. Converting the image to grayscale

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

# 5. Rotating and Shifting the image

```python
# Display both images side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Convert BGR to RGB for proper display with Matplotlib
rotated_image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_translation_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the rotated image on the left subplot
axes[0].imshow(rotated_image_rgb)
axes[0].set_title('Rotated Image')
axes[0].axis('off')

# Display the translated image on the right subplot
axes[1].imshow(img_translation_rgb)
axes[1].set_title('Translated Image')
axes[1].axis('off')

# Adjust layout for better visualization
plt.tight_layout()

# Show the plot
plt.show()
```
